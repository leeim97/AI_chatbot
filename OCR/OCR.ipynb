```python
import numpy as np
import platform
from PIL import ImageFont, ImageDraw, Image
from matplotlib import pyplot as plt
 
import uuid
import json
import time
import cv2
import requests
```

```python
api_url = ''
secret_key = ''
```

```python
# path = 'asset/images/test_image9.jpg'
path = 'C:/workspace_Multi07/OCR/data/성분표.png'
files = [('file', open(path,'rb'))]
```

images.url=> images.url 혹은 images.data 중 하나가 존재해야 함

```python

request_json = {'images': [{'format': 'jpeg',
                                'name': 'ingredient'
                               }],
                    'requestId': str(uuid.uuid4()),
                    'version': 'V2',
                    'timestamp': int(round(time.time() * 1000)),
                'lang':'ko'
                   }
```

```python
payload = {'message': json.dumps(request_json).encode('UTF-8')}
```

```python
headers = {
  'X-OCR-SECRET': secret_key,
}
```

```python
response = requests.request("POST", api_url, headers=headers, data=payload, files=files)
result = response.json()
```

```python
from langchain_openai import ChatOpenAI
from langchain.prompts import ChatPromptTemplate
from langchain_core.output_parsers import StrOutputParser
from langchain_community.document_loaders.csv_loader import CSVLoader
```

```python
# ans = """ https://www.10000recipe.com/recipe/’’’insert here’’’"""
# ans2 =  """{'name':’’’insert here’’’,'ingredients':’’’insert here’’’,'recipe':[’’’insert here’’’]}"""
# ans3 = """{'allergy':'''insert here''','food':'''insert here'''}"""


# Prompt
template = '''

Question : {question}

너는 성분을 알려주는 사람이야
{ing}를 보고 성분 요약해줘
'''

prompt = ChatPromptTemplate.from_template(template)

llm = ChatOpenAI(
    model='gpt-3.5-turbo-0125',
    temperature=0.6,
    max_tokens=850,
)

model_parser = llm | StrOutputParser()

query = '''성분 요약해서 알려줘'''

chain = (prompt 
          |  model_parser)
sum = chain.invoke({'question': query,'ing':ingred})
print(sum)
```
